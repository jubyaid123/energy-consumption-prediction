```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
predicted_weekly_output_meter1 = pd.read_csv('data/week_predictions.csv')
```

```python
actual_weekly_output = pd.read_excel('data/Evaluation Data_Shepard .xlsx')
```

```python
actual_weekly_output.head(10)
```

```python

actual_weekly_output.rename(columns={'CUNY City College of New York - 2023-09-01 -> 2023-09-30': 'interval_end_time'}, inplace=True)
actual_weekly_output.drop(actual_weekly_output.index[0], inplace=True)
actual_weekly_output.rename(columns={12785711: 'meter_1'}, inplace=True)
actual_weekly_output.rename(columns={'12785712': 'meter_2'}, inplace=True)
actual_weekly_output.set_index('interval_end_time', inplace=True)
actual_weekly_output.head(10)
```

```python
meter1_actual_output = actual_weekly_output[['meter_1']].copy()
meter2_actual_output = actual_weekly_output[['meter_2']].copy()
```

```python
plt.plot(actual_weekly_output.index, meter1_actual_output, label= 'meter 1 actual output')
plt.plot(actual_weekly_output.index, meter2_actual_output, label= 'meter 2 actual output')
plt.title('meter 1 and meter 2 actual output for first week of september')
plt.xlabel('Time')
plt.ylabel('Energy Output')
plt.legend()
plt.xticks(rotation=45)
plt.show()

```

```python
import matplotlib.dates as mdates
```

```python
meter1_actual_output.index
```

```python
# Convert timestamps to a format suitable for Matplotlib
timestamps = mdates.date2num(meter1_actual_output.index)

plt.figure(figsize=(15, 6))
plt.plot(timestamps, meter1_actual_output, label='Actual Energy Output')
plt.plot(timestamps, predicted_weekly_output_meter1, label='Predicted Energy Output', alpha=0.7)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())

plt.title('Comparison of Actual and Predicted Energy Output for meter one')
plt.xlabel('Time')
plt.ylabel('Energy Output')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```
